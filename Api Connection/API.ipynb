```python
import pandas as pd
import numpy as np
from pandas_datareader import data
from pytrends.request import TrendReq
```

```python
tickers = pd.read_excel("../Data/Market Ticker.xls")
tickers
```

```python
def yahoo_finance_data():
    TICKER = tickers["Ticker"]
    DATA_SOURCE = "yahoo"
    START = str(input("¿Desde que fecha desea obtener los datos diarios?--Formato usado: YYYY-MM-DD"))
    DATA_TABLE = data.DataReader(TICKER, DATA_SOURCE,START)
    return DATA_TABLE
```

```python
data_ent = yahoo_finance_data()
```

```python
data_ent
```

```python
last_data_end = data_ent.tail(10) #Cogemos las fechas de los 10 últimos días
```

```python
last_data_end
```

```python
last_data = last_data_end.transpose() #Data de los últimos 10 días
```
