```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../Data/Last_10_days_data_values.csv")
```

```python
data = data.set_index("Attributes")
data
```

```python
data = data.drop("Volume", axis = 0)
data = data.drop("Adj Close", axis = 0)
data = data.drop("Open", axis = 0)
data = data.drop("Close", axis = 0)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data = data.reset_index()
```

```python
data["Ticker"] = data[["Symbols", "Attributes"]].apply(lambda x: '_'.join(x), axis = 1)
```

```python
data = data.drop("Attributes", axis = 1)
data = data.drop("Symbols", axis = 1)
```

```python
data = data.set_index("Ticker")
data
```
